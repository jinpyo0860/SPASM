# Method 1-1. Transfering augmented data to Excel

```python
###11~50 or 13~60: Jittering
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sigma = 0.005

def DA_Jitter(X, sigma = 1E-4):
    myNoise = np.random.randint(-100,100, size=(241,1))*np.array(sigma)
    return X+myNoise

f_list = ['quick_dip','mid_dip','long_dip']

for i in f_list:
    counter = 0
    dir = '/content/gdrive/MyDrive/SPASM_project/3-class/SPASM_DA_Datasets/' + i + '/'
    file_list = os.listdir(dir)
    file_list_py = [file for file in file_list if file.endswith('xlsx')]

    if i == 'long_dip':
        for j in file_list_py:
            if int(j[0:-5]) <= 5:
                d1 = pd.read_excel(dir + j, header=None, index_col=None)

                for ii in range(8):
                    alpha = counter * 8
                    train_df = pd.DataFrame(DA_Jitter(d1, sigma))
                    writer = pd.ExcelWriter(dir + '%s.xlsx' % str(ii+11+alpha))
                    train_df.to_excel(writer, index = False, header=False)
                    writer.save()
                counter += 1
            else:
                pass
    else:
        for j in file_list_py:
              if int(j[0:-5]) <= 6:
                  d1 = pd.read_excel(dir + j, header=None, index_col=None)

                  for ii in range(8):
                      alpha = counter * 8
                      train_df = pd.DataFrame(DA_Jitter(d1, sigma))
                      writer = pd.ExcelWriter(dir + '%s.xlsx' % str(ii+13+alpha))
                      train_df.to_excel(writer, index = False, header=False)
                      writer.save()
                  counter += 1
              else:
                  pass
```

```python
###51~90 or 61~108: Scaling
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sigma = 0.1
def DA_Scaling(X, sigma=0.1):
    scalingFactor = np.random.normal(loc=1.0, scale=sigma, size=(1,X.shape[1]))
    myNoise = np.matmul(np.ones((X.shape[0],1)), scalingFactor)
    return X*myNoise

f_list = ['quick_dip','mid_dip','long_dip']

for i in f_list:
    counter = 0
    dir = '/content/gdrive/MyDrive/SPASM_project/3-class/SPASM_DA_Datasets/' + i + '/'
    file_list = os.listdir(dir)
    file_list_py = [file for file in file_list if file.endswith('xlsx')]

    if i == 'long_dip':
        for j in file_list_py:
            if int(j[0:-5]) <= 5:
                d1 = pd.read_excel(dir + j, header=None, index_col=None)
                d1 = d1.iloc[0:241,:]

                for ii in range(8):
                    alpha = counter * 8
                    train_df = pd.DataFrame(DA_Scaling(d1, sigma))
                    writer = pd.ExcelWriter(dir + '%s.xlsx' % str(ii+51+alpha))
                    train_df.to_excel(writer, index = False, header=False)
                    writer.save()
                counter += 1
            else:
                pass
    else:
        for j in file_list_py:
              if int(j[0:-5]) <= 6:
                  d1 = pd.read_excel(dir + j, header=None, index_col=None)
                  d1 = d1.iloc[0:241,:]

                  for ii in range(8):
                      alpha = counter * 8
                      train_df = pd.DataFrame(DA_Scaling(d1, sigma))
                      writer = pd.ExcelWriter(dir + '%s.xlsx' % str(ii+61+alpha))
                      train_df.to_excel(writer, index = False, header=False)
                      writer.save()
                  counter += 1
              else:
                  pass
```

```python
###91~130 or 109~156: Time warping
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline

def GenerateRandomCurves(X, sigma = 0.2, knot = 4):
    xx = (np.ones((X.shape[1],1))*(np.arange(0,X.shape[0], (X.shape[0]-1)/(knot+1)))).transpose()
    yy = np.random.normal(loc = 1.0, scale = sigma, size = (knot+2, X.shape[1]))
    x_range = np.arange(X.shape[0])
    cs_1 = CubicSpline(xx[:,0],yy[:,0])
    return np.array([cs_1(x_range)]).transpose()

def DistortTimesteps(X, sigma=0.2):
    tt = GenerateRandomCurves(X, sigma)
    tt_cum = np.cumsum(tt, axis=0)
    t_scale = [(X.shape[0]-1)/tt_cum[-1,0]]
    tt_cum[:,0] = tt_cum[:,0]*t_scale[0]
    return tt_cum

def DA_TimeWarp(X, sigma=0.2):
    tt_new = DistortTimesteps(X, sigma)
    X_new = np.zeros(X.shape)
    x_range = np.arange(X.shape[0])
    X_new[:,0] = np.interp(x_range, tt_new[:,0], X.iloc[:,0])
    return X_new

f_list = ['quick_dip','mid_dip','long_dip']

for i in f_list:
    counter = 0
    dir = '/content/gdrive/MyDrive/SPASM_project/3-class/SPASM_DA_Datasets/' + i + '/'
    file_list = os.listdir(dir)
    file_list_py = [file for file in file_list if file.endswith('xlsx')]

    if i == 'long_dip':
        for j in file_list_py:
            if int(j[0:-5]) <= 5:
                d1 = pd.read_excel(dir + j, header=None, index_col=None)
                d1 = d1.iloc[0:241,:]

                for ii in range(8):
                    alpha = counter * 8
                    train_df = pd.DataFrame(DA_TimeWarp(d1))
                    writer = pd.ExcelWriter(dir + '%s.xlsx' % str(ii+91+alpha))
                    train_df.to_excel(writer, index = False, header=False)
                    writer.save()
                counter += 1
            else:
                pass
    else:
        for j in file_list_py:
              if int(j[0:-5]) <= 6:
                  d1 = pd.read_excel(dir + j, header=None, index_col=None)
                  d1 = d1.iloc[0:241,:]

                  for ii in range(8):
                      alpha = counter * 8
                      train_df = pd.DataFrame(DA_TimeWarp(d1))
                      writer = pd.ExcelWriter(dir + '%s.xlsx' % str(ii+109+alpha))
                      train_df.to_excel(writer, index = False, header=False)
                      writer.save()
                  counter += 1
              else:
                  pass
```
