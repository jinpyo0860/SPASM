```python
import torch
import torch.nn.functional as F
import os
import pandas as pd
import numpy as np
#import rawdata as rd
from torch.utils.data import Dataset
from torch.utils.data import DataLoader


# Custom_Dataset

class CustomDataset_train(Dataset):
    # Data preprocessing part
    def __init__(self):
        self.x_data = x_train
        self.y_data = y_train

    # Length of the dataset (# of samples)
    def __len__(self):
        return len(self.x_data)

    # Get a certain sample in the dataset
    def __getitem__(self, idx):
        x = self.x_data[idx]
        y = self.y_data[idx]
        return x, y

class CustomDataset_test(Dataset):
    # Data preprocessing part
    def __init__(self):
        self.x_data = x_test
        self.y_data = y_test

    # Length of the dataset (# of samples)
    def __len__(self):
        return len(self.x_data)

    # Get a certain sample in the dataset
    def __getitem__(self, idx):
        x = self.x_data[idx]
        y = self.y_data[idx]
        return x, y

train_dataset = CustomDataset_train()
test_dataset = CustomDataset_test()
train_dataloader = DataLoader(train_dataset, batch_size=1, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=1, shuffle=False)
```

## Build CNN model

```python
################### 1d cnn model #######################################
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("다음 기기로 학습합니다:", device)
excel_data = pd.DataFrame(columns=["epoch","cost"])

# fix random seed (for reproducibility)
torch.manual_seed(777)

# fix random seed if GPU is available
if device == 'cuda':
    torch.cuda.manual_seed_all(777)

learning_rate = 0.001
training_epochs = 24

class CNN(nn.Module):

    def __init__(self):
        super(CNN, self).__init__()
        # 1st layer (CNN)
        self.layer1 = nn.Sequential(
            nn.Conv1d(1, 16, 3),
            nn.BatchNorm1d(16),
            nn.Tanh())

        # 2nd layer (CNN)
        self.layer2 = nn.Sequential(
            nn.Conv1d(16, 32, 3),
            nn.BatchNorm1d(32),
            nn.Tanh())

        # 3rd layer (CNN)
        self.layer3 = nn.Sequential(
            nn.Conv1d(32, 64, 3),
            nn.BatchNorm1d(64),
            nn.Tanh())

        # 4th layer (GAP)
        self.layer4 = nn.Sequential(
            nn.AdaptiveAvgPool1d(1),
            nn.Tanh())

        # 5th layer (FC)
        self.layer5 = nn.Sequential(
            nn.Linear(64, 3, bias=True),
            nn.Tanh())

    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = out.view(out.shape[0], -1)
        out = self.layer5(out)
        return out

# Defin the model
model = CNN().to(device)
```

```python
################################ Train & Test (loss & accuracy) #######################################
# Define cost function and optimizer
criterion = nn.CrossEntropyLoss().to(device) # softmax function is included in the cost function
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

total_batch_train = len(train_dataloader)
total_batch_test = len(test_dataloader)

print('train_num of total batch : {}'.format(total_batch_train))
print('test_num of total batch : {}'.format(total_batch_test))

for epoch in range(training_epochs + 1):

    # loss for trainsets
    avg_cost_train = 0

    for X_train, Y_train in train_dataloader:
        X_train = X_train.to(device)
        Y_train = Y_train.to(device)

        optimizer.zero_grad()
        hypothesis = model(X_train)
        cost = criterion(hypothesis, Y_train)
        cost.backward()
        optimizer.step()

        avg_cost_train += cost / total_batch_train

    # loss for testsets
    avg_cost_test = 0

    for X_test, Y_test in test_dataloader:
        X_test = X_test.to(device)
        Y_test = Y_test.to(device)

        optimizer.zero_grad()
        hypothesis = model(X_test)
        cost = criterion(hypothesis, Y_test)
        cost.backward()
        optimizer.step()

        avg_cost_test += cost / total_batch_test

    # accuracy for trainsets
    correct_train = 0
    total_train = 0

    with torch.no_grad():
        for data in train_dataloader:
            voltage_train, labels_train = data
            outputs = model(voltage_train)
            _, predicted_train = torch.max(outputs.data, 1)
            total_train += labels_train.size(0)
            correct_train += (predicted_train == labels_train).sum().item()

    # accuracy for testsets
    correct_test = 0
    total_test = 0

    with torch.no_grad():
        for data in test_dataloader:
            voltage_test, labels_test = data
            outputs = model(voltage_test)
            _, predicted_test = torch.max(outputs.data, 1)
            total_test += labels_test.size(0)
            correct_test += (predicted_test == labels_test).sum().item()

    print('[Epoch: {:>4}] Train_loss = {:>.9}'.format(epoch + 1, avg_cost_train) +
          ' Test_loss = {:>.9}'.format(avg_cost_test) +
          ' Train_acc: %f %%' % (100 * correct_train / total_train) +
          ' Test_acc: %f %%' % (100 * correct_test / total_test))

PATH = F"/content/gdrive/My Drive/SPASM_project/3-class/SPASM_DA_Datasets/Mwarping_ks=3_lr=0.001_epoch=25_tanh.pth"
torch.save(model.state_dict(), PATH)
```
